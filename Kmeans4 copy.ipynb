```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
xlsData = pd.read_excel("C:/workSpace(expert)/PM10.xlsx")
```

```python
data = xlsData[['Temp', 'Wind Speed']]
```

```python
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data)
```

```python
k = 3

model = KMeans(n_clusters = k, random_state = 10)
model.fit(data_scale)

xlsData['cluster'] = model.fit_predict(data_scale)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8, 8))

for i in range(k):
    plt.scatter(xlsData.loc[xlsData['cluster'] == i, 'Temp'], xlsData.loc[xlsData['cluster'] == i, 'Wind Speed'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel('Temp', size = 12)
plt.ylabel('Humidity', size = 12)
plt.show()
```
